### Sets(set)
- Unorderd , immutable and has unique elements (duplicates removed if added)
- example : set operations like union , intersection, storing membership
- so we can of set of tuples because they are immutable but we cannot create set of lists
- we can create them by using `{}` or using keyword `set`

Operation :
- Check if item is present using `in` keyword
- used `add()` to add items to the list
- use `remove()` to remove items from the list  but throws error if item is not present
- use `discard()` to remove items from the list  but `DOES NOT` throws error if item is not present
- Set operations
  -- union : combine two sets using `union()` or `|`
  -- inttersection : find command items between twos ets using `intersection()` or `&`
  --difference : find items in one set and not present in another set using `difference()` or `-`
  

```python
ports = set([80,443,8080,80,22])
print(ports)
names = ['priya','riya','tiya','siya']
print(names)
print(80 in ports)
ports.add(8081)
print(ports)
# ports.remove(82) # this will throw Key Error since the element is not present in set so use discard
ports.discard(82)
print(ports)
```

```python
#set_of_lists = set([[1,2],[3,6],[7,8]])  #TypeError: unhashable type: 'list' since we cannot create set of lists as they are mutable
# set_of_sets = {{ 1,2},{3,6},{7,8}} #TypeError: unhashable type: 'list' since we cannot create set of lists as they are mutable
set_of_tuples = {(1,3),(5.7),(13,17)}  # can create since tuples are immutable
print((1,2) in set_of_tuples)
```

```python
# set operations
developers = set(["alice","lad","tao"])
admins =set(["lee","parimal","lad"])
print("Union : ",developers.union(admins))
print("Intersection : ",developers.intersection(admins))
print("Difference : ",developers.difference(admins))
print("Union : ",developers | admins)
print("Intersection : ",developers & admins)
print("Difference : ",developers-admins)
```

- set of strings named required_packages and include some duplicates
- test for membership/presence of "aws" package in it
- add "terraform" and `safely` remove playwright from it
- create another set of  strings called installed . mention  some which are already present in required_packages
- On above two sets  compute missing ,  common and extra packages

```python
required_packages = set(['python3','playwright','pip','boto3','pip'])
print(required_packages)
print('aws' in required_packages)
print(required_packages.add('terraform'))
print(required_packages)
print(required_packages.discard('playwright'))
print(required_packages)
installed_packages = set(['python3','playwright','parser','synk','prettier'])
print("Missing packages : ",required_packages-installed_packages)
print("All packages : ",required_packages | installed_packages)
print("common packages : ",required_packages & installed_packages)
```
