- Indexing [] = access a character by index
- Slicing ([]) = Extracts sub strings
-  .lower() and .upper()

-  .startswith(), .endswith()
-  .split() , .join()

```python
title ="     My aspiration is to become a Cloud Automation Engineer    " 
print(title)
print(f"Result of len(): {len(title)}")
print(f"Result of upper(): {title.upper()}")
print(f"Result of lower(): {title.lower()}")
print(f"Result of .strip() : {title.strip()}")
print(f"Result of .lstrip() : {title.lstrip()}")
print(f"Result of .rstrip() : {title.rstrip()}")

filename ="file.yaml"
print("startes with file ? ",filename.startswith("file"))
print("ends with ml ? ",filename.endswith("ml"))


path = '/usr/local/bin'
path_parts = path.split('/')
print(f"path parts:{ path_parts}")
print(f"path parts joined : {('/').join(path_parts)}")

print(path + '/python')
print(path[3])
print(path[3:10])
print(path[3:])
print(path[:10])
```

### Sample Excercise

- Given the variables below, compute the disk usage percentage and print the raw percentage value.
- Build a human‑readable summary string:
    -  Convert the server name to uppercase.
    -  Include the number of CPU cores and amount of RAM.
    -  Show the disk usage percentage rounded to one decimal place.
- Print a summary containing the server name in uppercase, the number of CPU cores, the memory, and the disk usage).
- Finally, use the .2% format specifier in an f‑string to display the usage with two decimal places and a percent sign.


```python
server_name = "webserver-03"
cpu_cores = 4
memory_gb = 8.0
disk_total_gb = 500
disk_used_gb = 350

disk_used_percent = disk_used_gb/disk_total_gb
print(disk_used_percent)

summary = f"Server name : {server_name.upper()} CPU cores : {cpu_cores} RAM GB : {memory_gb} Disk usage : {disk_used_percent}"
print(summary)


summary_formatted = f"Server name : {server_name.upper()} CPU cores : {cpu_cores} RAM GB : {memory_gb} Disk usage : {disk_used_percent :.2%}"
print(summary_formatted)
```
