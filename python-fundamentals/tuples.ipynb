### Tuples (tuple)
- ordered but not mutable(immutable)
- once created we cannot change content
- they are defined within parenthesis ()

```python
host_port = ('127.0.0.1',3000)
rgb = (255,0,0)
single_tuple_value= ('single_value')
print(type(single_tuple_value))
print(type(rgb))
tuple_single_value = ('1 value',)  # because we add a training comma this will be considered as tuple
print(type(tuple_single_value))
```
