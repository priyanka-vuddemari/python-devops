```python
import math

print("When comparing floats directly , it might get precision issues")
print("0.1 * 3 == 0.3 :",0.1 * 3 == 0.3)
print("We can fix this precision issue by using math.isclose() method")
print("math.isclose(0.1 * 3, 0.3) :", math.isclose(0.1 * 3, 0.3))
```

### Arthimetic Operation 
- Supports  + , - , * , /
- /  -> evaluates to float
- // - floor division -> integer or  float
- % - modulo -> reminder
- ** - power

```python
print(8/2)
print(type(8/2))
print(5/3)
print(8//2)
print(type(8//2))
print(5//3)
print(5//3.0)
print(type(5//3.0))
print(6%4)
```
