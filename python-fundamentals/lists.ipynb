### List (list) is a mutable sequence of elements stored in square brackets 
- We can .append(), .insert(),  .pop() , .remove()
- Ideal for storing sequence where insertion order matters and contents change - deployemnt steps
- We can access the list using indexes. Use -ve for the last item
- We can get a slice from list by using myList[start:stop] to get from sublist from start and it will not include stop
- Three paramenter slicing mylist[start":stop:step] for steppigng for eg. mylist[::2] will give every other element
- If we do not give start,stop or then its defaults to begininng and ending - returns a new list without modifying the original

```python
servers = ["web01","web02","web03"]
mixed_list = ["config.yaml",123,True]

for item in mixed_list:
    print(type(item))

print(servers[0])
# print(servers[3]) # will throw an error as indes is out of range and IndexError Exception
print(servers[-1]) # return last item
print(servers[-2]) # returns last second item
print(servers[:2]) # returns before 2nd index means 0 to 1 
print(servers[1:]) # returns from 1st index means 1 to last 
print(servers[-2:]) # returns from last second element to last 
print(servers) # returns original
```

```python
# mutating list
ports =[80,443,8080]
ports.append(5000);  # adds elements at the end of list
print(ports)
ports.insert(1,3000) # add a element at particular index
print(ports)
# ports.remove? - get documentation about remove fucntion
ports.remove(443) # remove a element
print(ports)
removed_value = ports.pop(2) # removes a element at particular index
print(ports) 
print(removed_value)
```
